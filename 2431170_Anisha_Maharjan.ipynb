```python
# Define a function to categorize happiness based on the score
def categorize_happiness(score):
    if score < 4:  # If the score is less than 4, categorize as 'Low'
        return 'Low'
    elif 4 <= score <= 6:  # If the score is between 4 and 6 (inclusive), categorize as 'Medium'
        return 'Medium'
    else:  # If the score is greater than 6, categorize as 'High'
        return 'High'

# Apply the 'categorize_happiness' function to the 'score' column and create a new column 'Happiness Category'
df['Happiness Category'] = df['score'].apply(categorize_happiness)

# Display the first 10 rows of the DataFrame to view the new 'Happiness Category' column
df.head(10)
```

```python
# Define a list of South Asian countries to filter the data
south_asian_countries = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

# Filter the DataFrame 'df' to include only rows where the 'Country name' is in the South Asian countries list
south_asia_data = df[df["Country name"].isin(south_asian_countries)]

# Define the output file path where the filtered data will be saved
output_file_path = 'South_Asia_Happiness.csv'

# Save the filtered South Asian data to a CSV file without including the index column
south_asia_data.to_csv(output_file_path, index=False)
```

```python
# Calculate the Composite Score by applying weights to each feature (Log GDP per capita, Social support, and Healthy life expectancy)
south_asia_data['Composite Score'] = (
    0.40 * south_asia_data['Log GDP per capita'] +  # 40% weight for Log GDP per capita
    0.30 * south_asia_data['Social support'] +      # 30% weight for Social support
    0.30 * south_asia_data['Healthy life expectancy'] # 30% weight for Healthy life expectancy
)

# Sort the data based on the Composite Score in descending order (higher scores come first)
south_asia_data = south_asia_data.sort_values(by='Composite Score', ascending=False)

# Display the top 5 rows with the highest Composite Scores
south_asia_data.head(5)
```

```python
# Visualize the top 5 South Asian countries based on Composite Score using a horizontal bar chart
top_5_countries = south_asia_data.head(5)  # Select the top 5 countries
plt.figure(figsize=(10, 6))  # Set figure size
plt.barh(top_5_countries['Country name'], top_5_countries['Composite Score'], color='skyblue')  # Create a bar chart
plt.xlabel("Composite Score", fontsize=12)  # X-axis label
plt.ylabel("Country", fontsize=12)  # Y-axis label
plt.title("Top 5 South Asian Countries by Composite Score", fontsize=14)  # Plot title
plt.tight_layout()  # Adjust layout
plt.show()  # Display the chart
```

```python
# Plotting a comparison of Composite Score and Original Happiness Score for South Asian countries
plt.figure(figsize=(10, 6))  # Set figure size
plt.plot(south_asia_data['Country name'], south_asia_data['Composite Score'], label='Composite Score', marker='o')  # Composite Score
plt.plot(south_asia_data['Country name'], south_asia_data['score'], label='Original Score', marker='s')  # Original Score

# Adding labels, title, and legend
plt.xlabel("Country", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.title("Comparison of Composite Score and Original Score Rankings", fontsize=14)
plt.legend()
plt.tight_layout()  # Adjust layout for better visualization
plt.show()  # Display the plot

# Displaying the relevant data for better understanding
print(south_asia_data[['Country name', 'score', 'Composite Score']])
```

```python
# Handle missing values by dropping rows with missing values for simplicity in analysis
south_asia_df_cleaned = south_asia_data.dropna(subset=['score', 'Log GDP per capita'])

# Define outliers using the 1.5 * IQR rule
def identify_outliers(column):
    Q1 = south_asia_df_cleaned[column].quantile(0.25)
    Q3 = south_asia_df_cleaned[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return south_asia_df_cleaned[(south_asia_df_cleaned[column] < lower_bound) |
                                 (south_asia_df_cleaned[column] > upper_bound)]

# Identify outliers for 'score' and 'Log GDP per capita'
score_outliers = identify_outliers('score')
gdp_outliers = identify_outliers('Log GDP per capita')
```

```python
import seaborn as sns
# Scatter plot highlighting outliers
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=south_asia_df_cleaned, x='Log GDP per capita', y='score', label='South Asia Countries', s=100
)
sns.scatterplot(
    data=score_outliers, x='Log GDP per capita', y='score', color='red', label='Score Outliers', s=150
)
sns.scatterplot(
    data=gdp_outliers, x='Log GDP per capita', y='score', color='green', label='GDP Outliers', s=150
)
plt.title('Outliers in South Asia: Score vs GDP per Capita', fontsize=14)
plt.xlabel('Log GDP per Capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.legend()
plt.grid()
plt.show()

# Summary of outliers
score_outliers, gdp_outliers
```

```python
# Correlation analysis for selected metrics with Score
metrics = ['Freedom to make life choices', 'Generosity']
correlations = {
    metric: south_asia_df_cleaned[metric].corr(south_asia_df_cleaned['score']) for metric in metrics
}
```

```python
# Scatter plots with trendlines
plt.figure(figsize=(14, 6))
for i, metric in enumerate(metrics, 1):
    plt.subplot(1, 2, i)
    sns.regplot(
        data=south_asia_df_cleaned, x=metric, y='score',
        scatter_kws={'s': 100}, line_kws={'color': 'red'}
    )
    plt.title(f'Score vs {metric}', fontsize=14)
    plt.xlabel(metric, fontsize=12)
    plt.ylabel('Happiness Score', fontsize=12)
    plt.grid()

plt.tight_layout()
plt.show()

# Correlation results
correlations
```

```python
# Define Middle Eastern countries
middle_east_countries = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
    "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]

# Filter the dataset for Middle Eastern countries
middle_east_data = df[df['Country name'].isin(middle_east_countries)]

output_file_path1 = 'MiddleEast_Happiness.csv'
middle_east_data.to_csv(output_file_path, index=False)
```

```python
# 1. Descriptive Statistics
# Calculate mean and standard deviation of Score for South Asia and Middle East
south_asia_mean = south_asia_data['score'].mean()
south_asia_std = south_asia_data['score'].std()
middle_east_mean = middle_east_data['score'].mean()
middle_east_std = middle_east_data['score'].std()

print(f"South Asia - Mean Score: {south_asia_mean:.2f}, Std Dev: {south_asia_std:.2f}")
print(f"Middle East - Mean Score: {middle_east_mean:.2f}, Std Dev: {middle_east_std:.2f}")

# Which region has higher happiness scores on average?
if south_asia_mean > middle_east_mean:
    print("South Asia has a higher average happiness score.")
else:
    print("Middle East has a higher average happiness score.")
```

```python
# Top and bottom performers in each region
top_3_south_asia = south_asia_df_cleaned.nlargest(3, 'score')[['Country name', 'score']]
bottom_3_south_asia = south_asia_df_cleaned.nsmallest(3, 'score')[['Country name', 'score']]

top_3_middle_east = middle_east_data.nlargest(3, 'score')[['Country name', 'score']]
bottom_3_middle_east = middle_east_data.nsmallest(3, 'score')[['Country name', 'score']]

# Bar charts for top and bottom performers
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# South Asia
axes[0].bar(top_3_south_asia['Country name'], top_3_south_asia['score'], color='blue', label='Top 3')
axes[0].bar(bottom_3_south_asia['Country name'], bottom_3_south_asia['score'], color='orange', label='Bottom 3')
axes[0].set_title('South Asia: Top and Bottom Performers', fontsize=14)
axes[0].set_ylabel('Happiness Score', fontsize=12)
axes[0].legend()
axes[0].grid()

# Middle East
axes[1].bar(top_3_middle_east['Country name'], top_3_middle_east['score'], color='green', label='Top 3')
axes[1].bar(bottom_3_middle_east['Country name'], bottom_3_middle_east['score'], color='red', label='Bottom 3')
axes[1].set_title('Middle East: Top and Bottom Performers', fontsize=14)
axes[1].legend()
axes[1].grid()

plt.tight_layout()
plt.show()

# Displaying Top and Bottom Performers
top_3_south_asia, bottom_3_south_asia, top_3_middle_east, bottom_3_middle_east
```

```python
# Comparing key metrics: GDP per Capita, Social Support, and Healthy Life Expectancy
metrics = ['Log GDP per capita', 'Social support', 'Healthy life expectancy']

# Compute mean values for both regions
comparison_metrics = pd.DataFrame({
    'Metric': metrics,
    'South Asia': [south_asia_df_cleaned[metric].mean() for metric in metrics],
    'Middle East': [middle_east_data[metric].mean() for metric in metrics]
})

# Plot grouped bar chart for comparison
comparison_metrics_melted = comparison_metrics.melt(id_vars='Metric', var_name='Region', value_name='Mean Value')
plt.figure(figsize=(12, 6))
sns.barplot(data=comparison_metrics_melted, x='Metric', y='Mean Value', hue='Region')
plt.title('Comparison of Key Metrics Between South Asia and Middle East', fontsize=14)
plt.ylabel('Mean Value', fontsize=12)
plt.xlabel('Metric', fontsize=12)
plt.grid()
plt.show()

# Display metric comparison table
comparison_metrics
```

```python

# 4. Happiness Disparity
# Compute range and coefficient of variation for Score
south_asia_range = south_asia_data['score'].max() - south_asia_data['score'].min()
middle_east_range = middle_east_data['score'].max() - middle_east_data['score'].min()

south_asia_cv = south_asia_std / south_asia_mean
middle_east_cv = middle_east_std / middle_east_mean

print(f"South Asia - Range: {south_asia_range:.2f}, Coefficient of Variation: {south_asia_cv:.2f}")
print(f"Middle East - Range: {middle_east_range:.2f}, Coefficient of Variation: {middle_east_cv:.2f}")

# Which region has greater variability?
if south_asia_cv > middle_east_cv:
    print("South Asia has greater variability in happiness scores.")
else:
    print("Middle East has greater variability in happiness scores.")
```

```python
# 5. Correlation Analysis
# Correlation of Score with selected metrics
metrics_to_analyze = ['Freedom to make life choices', 'Generosity']
for metric in metrics_to_analyze:
    south_asia_corr = south_asia_data['score'].corr(south_asia_data[metric])
    middle_east_corr = middle_east_data['score'].corr(middle_east_data[metric])
    print(f"Correlation of Score with {metric} in South Asia: {south_asia_corr:.2f}")
    print(f"Correlation of Score with {metric} in Middle East: {middle_east_corr:.2f}")

# Scatter plots for correlations
for metric in metrics_to_analyze:
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.regplot(x=metric, y='score', data=south_asia_data)
    plt.title(f'South Asia: {metric} vs Score')
    plt.xlabel(metric)
    plt.ylabel('Score')

    plt.subplot(1, 2, 2)
    sns.regplot(x=metric, y='score', data=middle_east_data)
    plt.title(f'Middle East: {metric} vs Score')
    plt.xlabel(metric)
    plt.ylabel('Score')

    plt.tight_layout()
    plt.show()
```

```python
# Outlier detection for both regions
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

outliers_south_asia = detect_outliers(south_asia_df_cleaned, 'score')
outliers_middle_east = detect_outliers(middle_east_data, 'score')

# Scatter plots highlighting outliers for both regions
plt.figure(figsize=(14, 6))

# South Asia
plt.subplot(1, 2, 1)
sns.scatterplot(data=south_asia_df_cleaned, x='Log GDP per capita', y='score', label='South Asia', s=100)
sns.scatterplot(data=outliers_south_asia, x='Log GDP per capita', y='score', color='red', label='Outliers', s=150)
plt.title('South Asia: Score vs GDP (Outliers Highlighted)', fontsize=14)
plt.xlabel('Log GDP per Capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.legend()
plt.grid()

# Middle East
plt.subplot(1, 2, 2)
sns.scatterplot(data=middle_east_data, x='Log GDP per capita', y='score', label='Middle East', s=100)
sns.scatterplot(data=outliers_middle_east, x='Log GDP per capita', y='score', color='red', label='Outliers', s=150)
plt.title('Middle East: Score vs GDP (Outliers Highlighted)', fontsize=14)
plt.xlabel('Log GDP per Capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

# Display outliers for both regions
outliers_south_asia[['Country name', 'score', 'Log GDP per capita']], \
outliers_middle_east[['Country name', 'score', 'Log GDP per capita']]
```
